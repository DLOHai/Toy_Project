## Sapjil Issue
### (1) No more than 100

This library is not allowed to scrap more that 100 data at the same time. To deal with this problem, we should define the paths of `selenium` and `chromedriver`

### (2) Repeating File names 

Setting the argument of `no_numbering` as `True`

```
!pip install google_images_download 
from tqdm import tqdm
from google_images_download import google_images_download
```

```
names = ['트와이스 다현 얼굴', '다현 얼굴', '다현 face', 'dahyun face', 'twice dahyun face', 
         '레드벨벳 슬기 얼굴', '슬기 얼굴', '슬기 face','seulgi face', 'red velvet seulgi face']

def get_images(names):
    for name in tqdm(names):
        response = google_images_download.googleimagesdownload()
        arguments = {'keywords':name, 'limit':100, 'print_urls':False, 'no_numbering':True, 'output_directory':'./gdrive/My Drive/dlc_project1/'}   
        paths = response.download(arguments)

get_images(names)
```
